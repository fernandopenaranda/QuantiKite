```julia
using Quantica
using QuantiKite
```

```julia
function cubic_hamiltonian(; t = 1)
    σ0 = SA[1 0; 0 1]
    model = onsite((r) -> 2t * σ0) +
            hopping((r,dr) -> t/2 * σ0; range = 1)
    h = LP.cubic()|> supercell(1)|> hamiltonian(model, orbitals = 2) 
    return h
end
```

```julia
diskws = (OnsiteDisorderMeanStdv = 0.0, OnsiteDisorderMeanValue = 0.0, 
            OnsiteDisorderModelType = :gaussian,
            Concentration = 0.0)

c = configuration(
    divisions = (1,1,1), 
    unitcells =(16,16,16), 
    boundaries = ("periodic", "periodic", "periodic"),  
    precision = 2, # 
    spectrum_range = nothing
    );

h3D = cubic_hamiltonian()
```

#### DOS

```julia

s = dos(c, num_points = 10, num_moments = 10, num_random = 1, num_disorder = 1)
h5gen(h3D, c, s, filename = "dos_test.h5", disorder = true, vacancies = false; diskws...)
```
